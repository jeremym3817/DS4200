```python
import pandas as pd
import altair as alt

df = pd.read_csv("sample_superstore.csv", encoding="latin1", parse_dates=["Order Date"])

df_grouped = df.groupby(["Order Date", "Category"])["Sales"].sum().reset_index()

lineplot = alt.Chart(df_grouped).mark_line().encode(
    x="Order Date:T", 
    y="Sales:Q",      
    color="Category:N", 
    tooltip=["Order Date", "Category", "Sales"]
).properties(
    title="Sales Over Time by Category",
    width=800,
    height=400
)
lineplot
```

```python
import plotly.express as px

fig = px.treemap(df, 
                 path=['Region', 'State', 'Category', 'Sub-Category'],
                 values='Sales') 
fig.show()
```
